```python
sns.histplot(df["2023 population"],bins=10,kde=True)
plt.title("Distribution of 2023 Population")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby("continent")["2023 population"].mean()
```

```python
top_10_countries=df.sort_values(by="2023 population",ascending=False).head(10)
```

```python
sns.barplot(x="2023 population",y="country",hue="2023 population",data=top_10_countries,palette="viridis",legend=False)
plt.title("Top 10 Countries by Population in 2023")
plt.xlabel("Population")
plt.ylabel("Country")
plt.show()
```

```python
sns.scatterplot(x="2023 population",y="density (km²)",data=df)
plt.title("Population and Density Scatter Plot")
plt.xlabel("2023 Population")
plt.ylabel("Density")
plt.show()
```

```python
df.groupby("continent")["area (km²)"].sum().sort_values(ascending=False)
```

```python
sns.violinplot(x="continent",y="density (km²)",hue="continent",data=df,palette="viridis",legend=False)
plt.title("Population Density Distribution Across Continents")
plt.xlabel("Continent")
plt.ylabel("Populaton Density")
plt.show()
           
```

```python
sns.boxplot(x="continent",y="density (km²)",hue="continent",data=df,palette="muted",showfliers=False)
plt.title("Population Density Distribution Across Continents (Without Outliers)")
plt.xlabel("Continent")
plt.ylabel("Population Density")
plt.show()
```
